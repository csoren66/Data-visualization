Importing dataset

```python
# Seaborn for plotting and styling
import seaborn as sb
df = sb.load_dataset('tips')
print(df.head())
```

```python
#print all dataset available
print(sb.get_dataset_names())
```

Adding the third variable

```python
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
#basic scatter plot
sns.scatterplot(x="total_bill",
                y="tip",
                data=df)
plt.show()
```

```python
#scatter plot with hue (No order set)
sns.scatterplot(x="total_bill",
                y="tip",
                data=df,
                hue="smoker")
plt.show()
```

```python
#Setting hue order
sns.scatterplot(x="total_bill",
                y="tip",
                data=df,
                hue="smoker",hue_order=["Yes","No"])
plt.show()
```

```python
from seaborn import palettes
#Setting hue order and specifing hue color
hue_colors={"Yes":"black",
            "No":"red"}
sns.scatterplot(x="total_bill",
                y="tip",
                data=df,
                hue="smoker",hue_order=["Yes","No"],
                palette=hue_colors)
plt.show()
```

Using HTML hex color code with hue

```python
hue_colors={"Yes":"#808080",
            "No":"#00FF00"}
sns.scatterplot(x="total_bill",
                y="tip",
                data=df,
                hue="smoker",hue_order=["Yes","No"],
                palette=hue_colors)
plt.show()
```

Wrapping and ordering columns

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.relplot(x="total_bill",
                y="tip",
                data=df,
                kind="scatter",
                col="day",col_wrap=2,col_order=
                ["Thur","Fri","Sat","Sun"])
plt.show()
```

Subgroups with point size and hue

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.relplot(x="total_bill",
                y="tip",
                data=df,
                kind="scatter",
                size="size",
                hue="size")
plt.show()
```

```python
#Changing the transparacy(Set alpha between 0 and 1)
import matplotlib.pyplot as plt
import seaborn as sns

sns.relplot(x="total_bill",
                y="tip",
                data=df,
                kind="scatter",
                size="size",
                hue="size",
                alpha=0.4
                )
plt.show()
```

Line Plots in Seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns

#basic line plot
sns.relplot(x="total_bill",
                y="tip",
                data=df,
                kind="line")

#Subgroup by day by adding marker
sns.relplot(x="total_bill",
                y="tip",
                data=df,
                kind="line",style="day",hue="day",marker=True,
            dashes=False)
plt.show()
```

```python
#multiple observations per x values
#basic line plot
sns.relplot(x="day",
                y="total_bill",
                data=df,
                kind="scatter")
plt.show()
```

```python
sns.relplot(x="day",
                y="total_bill",
                data=df,
                kind="line")
plt.show()
```

```python
#replacing confidental interval with stdev
sns.relplot(x="day",
                y="total_bill",
                data=df,
                kind="line", ci='sd')
plt.show()
```

```python
# turn off confidential interval
sns.relplot(x="day",
                y="total_bill",
                data=df,
                kind="line",ci=None)
plt.show()
```

```python
#Count plot
import matplotlib.pyplot as plt
import seaborn as sns
sns.countplot(x="smoker",
              data=df)
plt.show()
```

Changing the order

```python
import matplotlib.pyplot as plt
import seaborn as sns

category_order=["No","Yes"]

sns.catplot(x="smoker",
              data=df, kind="count",
            order=category_order)
plt.show()
```

Divergent Palette

```python
from seaborn import categorical
sns.set_palette("RdBu")

category_order=["No","Yes"]

sns.catplot(x="smoker",
            data=df, kind="count",
            order=category_order)
plt.show()
```

```python
from seaborn import categorical
sns.set_palette("PRGn")

category_order=["No","Yes"]

sns.catplot(x="smoker",
            data=df, kind="count",
            order=category_order)
plt.show()
```

```python
#changing the scale
sns.set_context("paper")
sns.catplot(x="time",y="total_bill",data=df,kind="point")
plt.show()

sns.set_context("notebook")
sns.catplot(x="time",y="total_bill",data=df,kind="point")
plt.show()

sns.set_context("paper")
sns.catplot(x="time",y="total_bill",data=df,kind="point")
plt.show()

sns.set_context("talk")
sns.catplot(x="time",y="total_bill",data=df,kind="point")
plt.show()

sns.set_context("paper")
sns.catplot(x="time",y="total_bill",data=df,kind="point")
plt.show()
```

seaborn.boxplot

```python
import seaborn as sns
sns.set_theme(style="whitegrid")
tips = sns.load_dataset("tips")
ax = sns.boxplot(x=tips["total_bill"])
```

```python
ax = sns.boxplot(x="day", y="total_bill", data=tips)
```

```python
ax = sns.boxplot(x="day", y="total_bill", hue="smoker",
                 data=tips, palette="Set3")
```

```python
ax = sns.boxplot(x="day", y="total_bill", hue="time",
                 data=tips, linewidth=2.5)
```

```python
ax = sns.boxplot(x="time", y="tip", data=tips,
                 order=["Dinner", "Lunch"])
```

Draw a boxplot for each numeric variable in a DataFrame:

```python
iris = sns.load_dataset("iris")
ax = sns.boxplot(data=iris, orient="h", palette="Set2")
```

Use hue without changing box position or width:

```python
tips["weekend"] = tips["day"].isin(["Sat", "Sun"])
ax = sns.boxplot(x="day", y="total_bill", hue="weekend",
                 data=tips, dodge=False)
```

Use swarmplot() to show the datapoints on top of the boxes:

```python
ax = sns.boxplot(x="day", y="total_bill", data=tips)
ax = sns.swarmplot(x="day", y="total_bill", data=tips, color=".25")
```
