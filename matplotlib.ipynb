Visualizing & Customizing Plot appearances using Matplotlib

```python
# x-axis values
x = [5, 2, 9, 4, 7]

# y-axis values
y = [10, 5 , 8 , 4, 2]

# z-axis values
z = [9, 4 , 7 , 3, 1]
# Function to plot
plt.plot(x,y,marker="v", color="r", linestyle="--")
plt.plot(y,z,marker="o", color="g", linestyle="None")

# Function to show
plt.show()
```

```python
# x-axis values
x = [5, 2, 9, 4, 7]

# y-axis values
y = [10, 5 , 8 , 4, 2]

# Function to plot
plt.plot(x,y,marker="v", color="r", linestyle="--", label = "xy")
plt.plot(y,z,marker="o", color="b", linestyle="--", label = "yz")
plt.plot(z,x,marker="v", color="g", linestyle="--", label = "zx")

plt.xlabel(' x label')
plt.ylabel(' y label')
plt.title(" Sample plot")
plt.legend()

# Function to show
plt.show()
```

Multiple line plot

```python
plt.subplots()

fig, ax = plt.subplots(2, 1)

ax[0].plot(x,y,marker="v", color="r", linestyle="--")
ax[0].plot(y,z,marker="o", color="b", linestyle="--")

ax[0].set_xlabel(' x label')
ax[1].set_ylabel(' y label')
ax[1].set_xlabel(" Sample plot")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(10)
collectn_1=np.random.normal(100, 10, 200)
collectn_2=np.random.normal(80,30,200)
collectn_3=np.random.normal(90,20,200)
collectn_4=np.random.normal(70,25, 200)

## Combine the differnt collection into list
data_to_plot =[collectn_1, collectn_2, collectn_3,collectn_4]

# Create the figure instance
fig = plt.figure()

# Create an axes instance
ax=fig.add_axes([0,0,1,1])

#Create boxplot
bp=ax.violinplot(data_to_plot)
plt.show()
```

Contour plot

```python
import numpy as np
import matplotlib.pyplot as plt

xlist=np.linspace(-3.0,3.0,100)
ylist=np.linspace(-3.0,3.0,100)

X, Y = np.meshgrid(xlist, ylist)

Z = np.sqrt(X**2 + Y**2)

fig,ax =plt.subplots(1, 1)

cp=ax.contourf(X,Y,Z)
fig.colorbar(cp) # Add the color bar to the the plot

ax.set_title('Filled contour Plot')
ax.set_xlabel('x(cm)')
ax.set_ylabel('y(cm)')

plt.show()
```

Annotations

```python
# Implementation of matplotlib.pyplot.annotate()
# function
  
import matplotlib.pyplot as plt
import numpy as np
  
  
fig, geeeks = plt.subplots()
  
t = np.arange(0.0, 5.0, 0.001)
s = np.cos(3 * np.pi * t)
line = geeeks.plot(t, s, lw = 2)
  
# Annotation
geeeks.annotate('Local Max', xy =(3.3, 1),
                xytext =(3, 1.8), 
                arrowprops = dict(facecolor ='green',
                                  shrink = 0.05),)
  
geeeks.set_ylim(-2, 2)
  
# Plot the Annotation in the graph
plt.show()
```

Advance annotions

1. Sine waveform

```python
import matplotlib.pyplot as plt
import numpy as pp
 
fig, ppool = plt.subplots()
 
t = pp.arange(0.0, 1.0, 0.001)
s = pp.sin(2 * pp.pi * t)
line = ppool.plot(t, s, lw=2)
 
ppool.annotate('Peak Value', xy=(.25, 1),
                xytext=(1, 1),arrowprops=dict(facecolor='yellow',
                                shrink=0.05),xycoords="data",)
 
ppool.set_ylim(-1.5, 1.5)
plt.show()
```

1. Annotate Scatter Plot

```python
import matplotlib.pyplot as plt
y = [3.2, 3.9, 3.7, 3.5, 3.02199]
x = [0.15, 0.3, 0.45, 0.6, 0.75]
n = [155, "outliner", 293, 230, 670]
 
fig, ax = plt.subplots()
ax.scatter(x, y)
 
for i, txt in enumerate(n):
    ax.annotate(txt, (x[i], y[i]))
plt.show()
```
